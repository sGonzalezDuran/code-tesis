```python
import os
import pandas as pd

ruta_carpeta = 'images/belenegri'

data = {'carpeta_numero': [], 'nombres_imagenes': []}

for raiz, carpetas_directorio, archivos in os.walk(ruta_carpeta):
    nombres_imagenes = []
    for archivo in archivos:
        extensiones_permitidas = [".jpg", ".jpeg", ".png", ".gif"]
        if any(archivo.endswith(ext) for ext in extensiones_permitidas):
            nombres_imagenes.append(archivo)

    if nombres_imagenes:
        carpeta_numero = os.path.basename(raiz)
        data['carpeta_numero'].append(carpeta_numero)
        data['nombres_imagenes'].append(nombres_imagenes)

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
